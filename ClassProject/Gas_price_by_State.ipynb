```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#prepare data 
gasPrices = pd.read_csv('./gas_price_by_state.csv')
#print(gasPrices)

labels = gasPrices['name']
values = gasPrices['gasoline']

plt.figure(figsize=(15, 4))
plt.xticks(rotation=90)
plt.title('Gas Prices per State', fontdict={'fontname':'Comic Sans MS', 'fontweight': 'bold'})
plt.xlabel('State', fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Gasoline Price', fontdict={'fontname':'Comic Sans MS'})
plt.yticks([3.30, 4.00, 4.50, 5.00, 5.50])
plt.ylim(3, 5.5)

gasBar = plt.bar(labels, values)
plt.show()
```
