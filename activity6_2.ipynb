XML

3. Here is the url to an example xml file: https://www.w3schools.com/xml/cd_catalog.xml Use the requests library to download the page and parse the xml file.

```python
# Your Code Here (imports done for you)
import xml.etree.ElementTree as ET
import urllib.request, urllib.parse, urllib.error
import pandas as pd

url = 'https://www.w3schools.com/xml/cd_catalog.xml'
response = urllib.request.urlopen(url).read()
#print(response)
tree = ET.fromstring(response)


#print(root)
for cd in tree.findall('CD'):
    title = cd.find('TITLE').text
    artist = cd.find('ARTIST').text
    print("TITLE:", title, 'ARTIST:', artist)

```
